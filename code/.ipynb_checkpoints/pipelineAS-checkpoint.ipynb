```python
%autoreload
t_i = time.time()
waveforms = crc.add_A_phi(crc.load_waveforms(scenario, n, mode))
print(f"\n{time.time() - t_i:.3f} seconds")
print(len(waveforms))
```

```python
%autoreload
del_params = []
del_keys = []
guide = list(waveforms.keys())
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
ax.set_ylabel(fr"$A_{{{mode.replace(',', '')}}}^{{\text{{{scenario}}}}}(t)$")
ax.set_xlabel(fr"$t (t/M)$")
for i, q in enumerate(guide):
    A = waveforms[q]['wf']['A']
    plt.plot(crc.GET_TIME(), A, c = 'k', alpha = 0.1, zorder = 0)
    if np.argmin(A[:5000+1]) != 0:
        del_params.append(f"h_{i}.npy")
        del_keys.append(q)
        plt.plot(crc.GET_TIME(), A, c = 'r')
        del waveforms[q]
        continue
    if np.argmax(A[:2500+1]) != 2500:
        del_params.append(f"h_{i}.npy")
        del_keys.append(q)
        plt.plot(crc.GET_TIME(), A, c = 'r')
        # del waveforms[q]
        continue
    if np.abs(A[10] - A[0]) > np.abs(A[110] - A[100]):
        del_params.append(f"h_{i}.npy")
        del_keys.append(q)
        plt.plot(crc.GET_TIME(), A, c = 'r')
        # del waveforms[q]
        continue
# print(del_params)
plt.tight_layout()
plt.savefig(f"../figures/{scenario}/bad_wf_{mode.replace(',', '')}")
```
