```python
import numpy as np
import crc
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
rc_params = {
    'axes.grid'             : False,
    'axes.labelsize'        : 24,
    'axes.linewidth'        : 1,
    'axes.titlesize'        : 24,
    'font.size'             : 24,
    'legend.fontsize'       : 16,
    'xtick.labelsize'       : 12,
    'ytick.labelsize'       : 16,
    # 'font.family'           :'sans-serif',
    'font.family'           :'serif',
    'text.latex.preamble'   : r'\usepackage{amsmath} \usepackage{amssymb} \usepackage{amsfonts}',
    'text.usetex'           : False,
    'patch.force_edgecolor' : True,
    'figure.dpi'            : 256,
    'mathtext.fontset'      : 'cm'     
}

mpl.rcParams.update(rc_params)
```

```python
x = np.linspace(-1, 1, 1001)
ns = range(6)
V = {}
for n in ns:
    V.update({n : x**n})
U = crc.MGS(V, x)
# crc.plot_rb(U, x, Legendre = True)
```

```python
cL = crc.sort_hex(crc.get_colours(16))[:6]
crc.colour_fam(cL)
```

```python
cL2 = crc.sort_hex(crc.get_colours(12))[:6]
crc.colour_fam(cL2)
```

```python
c0 = crc.get_colours(len(U))
fig, ax = plt.subplots(1, 1, figsize = (10, 6));
for n, i in enumerate(U):
    ax.plot(x, U[i] / np.sqrt(i + 1/2), color = cL2[i], label = fr"$P_{n}(x)$");
ax.set_xlabel(fr"$x$");
ax.set_ylabel(fr"$P_n(x)$")
plt.legend();
fig.tight_layout()
plt.savefig(f"../figures/mgs_legendre_polynomials")
```
