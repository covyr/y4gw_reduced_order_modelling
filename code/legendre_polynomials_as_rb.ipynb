```python
import numpy as np
import crc
import matplotlib.pyplot as plt
import time
```

```python
x = np.linspace(-1, 1, 1001)
V = {n: x ** n for n in range(1000)}
tolerance = 1e-8
```

```python
t_i = time.time()
rb, errors, indexes = crc.RB(V, x, tolerance)
print(f"\n{time.time() - t_i:.3f} seconds")
crc.plot_greedy(errors, tolerance, None)
print("reduced basis:")
crc.plot_rb(rb, x, Legendre = True)
print("waveforms associated with the rb elements:")
crc.plot_rb({n: x ** n for n in rb}, x, Legendre = False)
```
