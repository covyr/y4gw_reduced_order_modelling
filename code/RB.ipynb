```python
import numpy as np
import crc
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 1001)
vectors = [{n : x**n for n in range(1000)}]
V = {}
for vec in vectors:
    V.update(vec)
```

```python
rb, errors, indexes = crc.RB(V, x, 1e-8)
```

```python
crc.plot_greedy(errors, 1e-8, None)
```

```python
crc.plot_rb(rb, x, True)
```
